```python
import numpy as np
```

```python
# 1D Array
arr1 = np.array([1, 2, 3, 4])
print("1D Array:", arr1)

# 2D Array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

# 3D Array
arr3 = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print("3D Array:\n", arr3)
```

```python
arr = np.array([10, 20, 30, 40, 50])

# Indexing
print("First Element:", arr[0])

# Slicing
print("Slice:", arr[1:4])

# Reshaping
arr_reshape = np.array([1,2,3,4,5,6])
reshaped = arr_reshape.reshape(2,3)

print("Reshaped Array:\n", reshaped)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

# Element-wise operations
print("Addition:", a + b)
print("Multiplication:", a * b)

# Broadcasting
c = np.array([10])
print("Broadcasting:", a + c)
```

```python
data = np.array([10, 20, 30, 40, 50])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Percentile (25%):", np.percentile(data, 25))
```

```python
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])

# Dot Product
print("Dot Product:\n", np.dot(A, B))

# Matrix Multiplication
print("Matrix Multiplication:\n", A @ B)
```

```python
# Random numbers
rand_nums = np.random.rand(5)
print("Random Numbers:", rand_nums)

# Random integers
rand_ints = np.random.randint(1, 100, size=5)
print("Random Integers:", rand_ints)

# Random sample from array
data = np.array([10,20,30,40,50])
sample = np.random.choice(data, size=3)

print("Random Sample:", sample)
```

```python
# Example stock prices
prices = np.array([100, 102, 101, 105, 110, 108])

# Daily returns
returns = np.diff(prices)

print("Prices:", prices)
print("Daily Changes:", returns)

print("Average Price:", np.mean(prices))
print("Max Price:", np.max(prices))
print("Min Price:", np.min(prices))
```
