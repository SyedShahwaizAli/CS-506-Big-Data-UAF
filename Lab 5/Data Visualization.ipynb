```python
# Practical Tasks 5 - Data Visualization Portfolio

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (COVID example CSV)
df = pd.read_csv('covid_data.csv')

# Convert date column
df['date'] = pd.to_datetime(df['date'])

# 30. Line Plot (Time Series)
plt.figure()
plt.plot(df['date'], df['cases'])
plt.title('COVID Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.show()

# 31. Bar Chart
plt.figure()
df.groupby('country')['cases'].sum().plot(kind='bar')
plt.title('Total Cases by Country')
plt.show()

# Histogram
plt.figure()
plt.hist(df['cases'], bins=30)
plt.title('Distribution of Cases')
plt.show()

# 32. Scatter Plot
plt.figure()
plt.scatter(df['cases'], df['deaths'])
plt.xlabel('Cases')
plt.ylabel('Deaths')
plt.title('Cases vs Deaths')
plt.show()

# 33. Box Plot
plt.figure()
sns.boxplot(x=df['cases'])
plt.title('Box Plot of Cases')
plt.show()

# Violin Plot
plt.figure()
sns.violinplot(x=df['cases'])
plt.title('Violin Plot of Cases')
plt.show()

# 34. Heatmap (Correlation)
plt.figure()
corr = df[['cases', 'deaths']].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()

# 35. Customized Plot
plt.figure()
plt.plot(df['date'], df['cases'], label='Cases')
plt.plot(df['date'], df['deaths'], label='Deaths')
plt.legend()
plt.title('Cases vs Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()

# 36. Subplots
fig, axes = plt.subplots(2, 2)

axes[0,0].plot(df['date'], df['cases'])
axes[0,0].set_title('Cases')

axes[0,1].plot(df['date'], df['deaths'])
axes[0,1].set_title('Deaths')

axes[1,0].hist(df['cases'])
axes[1,0].set_title('Cases Histogram')

axes[1,1].scatter(df['cases'], df['deaths'])
axes[1,1].set_title('Scatter')

plt.tight_layout()
plt.show()

print("Visualization portfolio complete!")
```
